```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
DataListOne = [x for x in range(0,20,1)]

#DateList = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2000', periods=1000))

DataFr = pd.DataFrame(data = {'Name': ['Tom', 'Joseph', 'Krish', 'John'], 'Age': [20, 21, 19, 18]})
```

```python
DataFr
```

```python
plt.plot(DataListOne)
```

```python
plt.figure(figsize=(8,6), dpi= 92)
plt.xlabel('Names',fontsize = 20)
plt.ylabel('Age',fontsize = 20)
plt.grid()
plt.title('Name VS. Age',fontsize = 20)
plt.bar(x=DataFr['Name'], data=DataFr['Age'], height=DataFr['Age'])
plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
Name = ['Salary', 'Stocks', 'Freelance', 'Others']
Presint = ['68.2', '1.8', '19.5', '10.5']
#colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
#plt.title('Sorces of Income Monthly',fontsize = 20)
plt.pie(Presint, labels=Name, autopct='%1.1f%%',  startangle=180, radius=1.5)

plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.figure(dpi=96)
plt.scatter(DataFr['Name'],DataFr['Age'], alpha = 0.4, s=50)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.figure(dpi=96)
plt.scatter(DataFr['Name'],DataFr['Age'], alpha = 0.4, s=50)
plt.title('Age vs. Name',fontsize = 20, loc='left', color='red')
plt.xlabel('Names',fontsize = 16)
plt.ylabel('Ages',fontsize = 16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
Lis1 = [x for x in np.random.rand(20) + np.random.rand(20)]
Lis2 = [x for x in np.random.rand(20) + np.random.rand(20)]
Lis3 = [x for x in np.random.rand(20) + np.random.rand(20)]

plt.figure(dpi=96)
plt.plot(Lis1)
plt.plot(Lis2)
plt.plot(Lis3)
plt.legend(['rand1', 'Rand2', 'Rand3'])
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(1,3,1)
plt.pie(Presint, labels=Name, radius=1.5)
fig.add_subplot(1,3,2)
plt.scatter(DataFr['Name'],DataFr['Age'], alpha = 0.4, s=50)
plt.title('Age vs. Name',fontsize = 20, loc='left', color='red')
plt.xlabel('Names',fontsize = 16)
plt.ylabel('Ages',fontsize = 16)
fig.add_subplot(1,3,3)
plt.plot(Lis1)
plt.plot(Lis2)
plt.plot(Lis3)
plt.legend(['rand1', 'Rand2', 'Rand3'])
plt.show()
```
